```python
#final run
import os
import csv
import pandas as pd

# Specify the parent folder containing subfolders with CSVs
parent_folder = "1th_scene-obj"

# Iterate through subfolders
for subfolder in os.listdir(parent_folder):
    subfolder_path = os.path.join(parent_folder, subfolder)
    print(subfolder_path)
    
    # Check if it's a directory
    if os.path.isdir(subfolder_path):
        # Code for deleting specified columns
        columns_to_delete = ['time', '.id', '.f_VrelX', '.f_VrelY', '.f_ArelX', '.f_ArelY', '.f_VrelXStd', '.f_VrelYStd', '.f_ArelXStd', '.f_ArelYStd', '.f_LDeltaX_left', '.f_LDeltaX_mid', '.f_LDeltaX_right', '.f_ObjectScore', '.f_Orientation', '.f_OrientationStd', '.f_DistZ', '.f_DistZStd', 'timestamps', '.f_LDeltaY_left', '.f_LDeltaY_mid', '.f_LDeltaY_right']
        
        for filename in os.listdir(subfolder_path):
            if filename.endswith('.csv'):
                file_path = os.path.join(subfolder_path, filename)
                
                # Read the CSV file
                df = pd.read_csv(file_path)
                
                # Delete the specified columns
                df.drop(columns=columns_to_delete, inplace=True)
                
                # Save the modified dataframe back to the CSV file
                df.to_csv(file_path, index=False)
        
        # Code for transposing CSV files
        for file_name in os.listdir(subfolder_path):
            if file_name.endswith(".csv"):
                input_file_path = os.path.join(subfolder_path, file_name)
                print(input_file_path)
                # Read the CSV file
                with open(input_file_path, 'r') as csv_file:
                    reader = csv.reader(csv_file)
                    data = list(reader)
                
                # Transpose the data
                transposed_data = list(map(list, zip(*data)))
                
                # Write the transposed data back to the same CSV file
                with open(input_file_path, 'w', newline='') as csv_file:
                    writer = csv.writer(csv_file)
                    writer.writerows(transposed_data)
        
        # Code for modifying the second column
        for file_name in os.listdir(subfolder_path):
            if file_name.endswith(".csv"):
                input_file_path = os.path.join(subfolder_path, file_name)
                
                with open(input_file_path, 'r') as input_file:
                    reader = csv.reader(input_file)
                    rows = list(reader)
                    
                    for row in rows:
                        if len(row) >= 2:
                            row[1] = row[1].lstrip('(').rstrip(')')
                    
                with open(input_file_path, 'w', newline='') as output_file:
                    writer = csv.writer(output_file)
                    writer.writerows(rows)
        
        # Code for adding labels and values to CSV files
        for file_name in os.listdir(subfolder_path):
            if file_name.endswith(".csv"):
                input_file_path = os.path.join(subfolder_path, file_name)
                
                new_row = ['Parameters', 'Values']
                
                existing_rows = []
                with open(input_file_path, 'r') as file:
                    reader = csv.reader(file)
                    existing_rows = list(reader)
                
                updated_rows = [new_row] + existing_rows
                
                with open(input_file_path, 'w', newline='') as file:
                    writer = csv.writer(file)
                    writer.writerows(updated_rows)
        
        # Code for extracting values from the second column
        for file_name in os.listdir(subfolder_path):
            if file_name.endswith(".csv"):
                input_file_path = os.path.join(subfolder_path, file_name)
                
                df = pd.read_csv(input_file_path)
                second_column = df.iloc[:, 1]
                values = second_column.tolist()
                
                with open(input_file_path, 'w') as file:
                    for value in values:
                        file.write(str(value))
                        file.write('\n')
```
