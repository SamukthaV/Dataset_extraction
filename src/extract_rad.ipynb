```python
import pandas as pd
import os
temp = [25]
j=1
for i in temp:
    filet = '/home/rad-cam/solio_dataset/extract_radar/solio_lrr/'+str(i)+'-scene'+str(1)+'-radar_obj.csv'
    print(filet)
    df = pd.read_csv(filet)
    print(df)
#add new column to edit the time column
    df['timestamps'] = df['time'].str.replace('2023/09/02/', '')

    inputf = str(j)+'th_scene-obj/'+str(i)
    input_dir = inputf

    if not os.path.exists(input_dir):
        os.makedirs(input_dir)
    # Assuming that df contains the data you need

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        # Extract 'timestamps' value for the row
        timestamps = row['timestamps']

        # Replace ':' with '-' in timestamps
        timestamps = timestamps.replace(':', '-')
        timestamps = timestamps.replace('/', '_')

        # Generate the filename using 'timestamps'
        filename = os.path.join(input_dir, f'{timestamps}.csv')
        # print("dddd: ",filename)

        # Create a new DataFrame for the row
        row_df = pd.DataFrame(row).transpose()

        # # Save the DataFrame as a CSV file
        row_df.to_csv(filename, index=False)
```
